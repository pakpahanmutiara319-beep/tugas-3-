```python
def power(x, n):
    if n == 0:
        return 1
    
    if n < 0:
        return 1 / power(x, -n)
    
    return x * power(x, n - 1)

try:
    x = float(input("Masukkan nilai x: "))
    n = int(input("Masukkan pangkat n: "))
    
    hasil = power(x, n)
    print("Hasil:", hasil)

except ValueError:
    print("Input harus angka!")
```
